**a.** Given that 

```python

```

using **ONLY** `NumPy`, write an equation that calculates `y`. `y` is calculated using the equation below:
 
$$
y = \exp^{(\frac{-x}{10})}*\sin{(x)}
$$

<br>

**b.** Write a NumPy code that calculates: 

$$
result = \frac{\sum{((x - \bar{x}) * (y - \bar{y}))}}{\sum{(x - \bar{x})^2}}
$$


`X` and `y` are given below:

```python
rng = np.random.default_rng(seed=5)
X = rng.random(size=(35, 2), dtype=np.float32)
y = rng.random(size=(35, 1), dtype=np.float32)
```

````python
# Write your solution here
# Qs 1a
x = np.linspace(0,10,10001)
y = (np.exp(-abs(x)/10)) *  np.sin(x)
y
````

````python
# Write your solution here
# Qs 1b
rng = np.random.default_rng(seed=5)
x = rng.random(size=(35,2), dtype=np.float32)
y = rng.random(size=(35,1), dtype=np.float32)

numerator = np.sum((x-np.mean(x))*(y-np.mean(y)))
denomenator = np.sum((x-np.mean(x))**2)
result = numerator/denomenator
result
````

**a.** Create a `DataFrame` using the lists above. The column names should be: `name` , `job_role` and `salary`.

**b.** Create a new column **`title`** that has has 2 new labels `Engineer` and `Analyst/Scientist` if the person has `engineer` or `analyst/scientist` in their job role.

**c.** What is the mean and median salary per `title`?

````python
# Write your solution here
# Qs 4a

salary = [74454, 54900, 76182, 56221, 81254, 75900, 67122, 62426, 61553, 77070, 67587, 50095, 75820, 68174, 79069]
students = ["Sadiq","Aminu","Udensi","Olayiwola","Ukpeh","Akinbade","Ezeofor","Oludele","Gbenga","Edgar","Worlu",
            "Magbuwe","Atte","Orogun","Bamigbade"]
roles = ["Data Engineer","Data Scientist","DevOps Engineer","Research Engineer","Analytical Engineer","Data Analyst",
         "ML Engineer","DevOps Engineer","Analytical Engineer","Data Scientist","MLOps Engineer","Data Engineer",
         "Data Scientist","ML Engineer","Research Engineer",]

df = pd.DataFrame({'name':students, 'job_role':roles, 'salary':salary})
df
````

````python
# Write your solution here
# Qs 4b
def get_title(row:str) -> str:
    if pd.Series(row['job_role']).str.lower().str.contains('engineer').any():
        role = 'Engineer'
    elif pd.Series(row['job_role']).str.lower().str.contains('analyst|scientist').any():
        role = 'Analyst/Scientist'
    return role
df['title'] = df.apply(lambda row: get_title(row), axis=1)
df
````

````python
# Write your solution here
# Qs 4c
df.groupby('title')['salary'].agg([np.mean,np.median])
````

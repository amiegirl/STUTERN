### Question 1
Given the following dataset:

a = np.random.normal(loc=64, scale=2, size=500) \
b = np.random.normal(loc=70, scale=2, size=100)

Write the appropriate to display multiple histogram on the same axis. The plots for both datasets show be fully
visible.
[Hint: Use the 'alpha' argument]

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#1
a = np.random.normal(loc=64, scale=2, size=500)
b = np.random.normal(loc=70, scale=2, size=100)

# Histogram of data
plt.figure(figsize =[6,4])
np.random.seed(0)
plt.hist(a, bins=5)
plt.hist(b, bins=5, color='green', alpha = 0.8)
plt.title('Multiple Histogram of normal variable')
plt.xlabel('a vs b')
plt.ylabel('Frequencies');
```

### QUESTION 3:
How would you use matplotlib module to make two different subplots on different axes with the following 
characteristics?

a) First subplot: \
i. pie chart using the 'sex' column. Ensure the percentage on the pie chart is taken to 2 d.p.

b) Second subplot: \
i. Create a bar-chart with the 'region' column. \
ii. Carefully label the chart with x-axis: 'Region'; y-axis: 'Counts'; title: 'Bar plot for Region'.


```python
#3
# pie chart
fig,(ax1,ax2)= plt.subplots(1, 2, figsize=[8,3])
gender_counts= df['sex'].value_counts()
ax1.pie(gender_counts, labels=gender_counts.index, autopct='%1.2f%%')
ax1.set_title('Gender Proportion')

# bar chart
x= df['region'].unique()
y= df['region'].value_counts()
ax2.bar(x,y)
ax2.set_xlabel('Region')
ax2.set_ylabel('Counts')
ax2.set_title('Bar plot for Region')
plt.subplots_adjust(wspace= 0.5);
ax2.tick_params('x', labelrotation=45)
```

### QUESTION 4:

Using matplotlib, plot two histograms on the same axis for columns 'age' and 'children'. Ensure the following
is indicated in the plot: 
1. The two plots should be clearly visible to the eye [Hint: Use the 'alpha' argument] \
ii. Use appropriate titles for the axes and the plot

```python
plt.figure(figsize =[6,3])
plt.hist(df['age'], bins=10, label='age')
plt.hist(df['children'], bins=10, label= 'children', alpha = 0.8)
plt.title('Plot of Age & Children')
plt.xlabel('Age & Children')
plt.ylabel('Count')
plt.legend();
```

### QUESTION 5:

Using the dataset given below,

x_var = np.random.normal(loc = 40, scale = 1.5, size = 200) \
y_var = np.random.normal(loc = 50, scale = 1.5, size = 300) \
z_var = np.random.normal(loc = 60, scale = 2, size = 300)

i. Create histograms for the two variables,x_var and y_var, on the same axis. Ensure the two plots can easilybe see using the appropriate parameter. Also, use the legend() function for the plot.

```python
#i
x_var = np.random.normal(loc = 40, scale = 1.5, size = 200) 
y_var = np.random.normal(loc = 50, scale = 1.5, size = 300)
z_var = np.random.normal(loc = 60, scale = 2, size = 300)
np.random.seed(0)
plt.figure(figsize=[6,3])
plt.hist(x_var, bins=10, label='x_var')
plt.hist(y_var, bins=10, alpha = 0.8, color='green', label='y_var')
plt.title(' Multiple histogram: x_var & y_var')
plt.xlabel('x_var vs y_var')
plt.ylabel('Counts')
plt.legend();
```

ii. Make two subplots using the 3 variables as follows:

1. Make a lineplot of y_var against z_var for the first subplot
2. Make a histogram for the second subplot using x_var.
[Hint: All plots should be well labeled]

```python
#ii
fig,(ax1,ax2)= plt.subplots(1, 2, figsize=[6,3])
ax1.plot(y_var, z_var, color='blue')
ax1.set_xlabel('y_var')
ax1.set_ylabel('z_var')
ax1.set_title('Lineplot: y_var vs z_var')
ax2.hist(x_var, bins= 10)
ax2.set_xlabel('x_var')
ax2.set_ylabel('Counts')
ax2.set_title('Histogram of x_var')
plt.subplots_adjust(wspace= 0.5);
```
